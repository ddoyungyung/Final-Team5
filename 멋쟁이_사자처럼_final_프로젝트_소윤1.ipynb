```
특이한 변수: Negative_Review
"If the reviewer did not write a negative review, it is represented as "No Negative"

따라서, 리뷰어가 긍정적인 리뷰만 남기고 부정적인 리뷰칸을 비워두었다(부정적인 리뷰를 작성하지 않았다) == No Negative로 표시!
```

````python
import re
import string

def clean_text(text):
    if not isinstance(text, str): # 문자열이 아닐 경우 빈 문자열 반환
        return ''

    if 'no negative' in text.lower():
        return ''
    text = re.sub(r'\[.*?\]', '', text) #re.sub : 정규표현식을 이용해 특정 패턴의 문자열을 다른 문자열로 대체
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    #string.punctuation : string라이브러리에 미리 정의된 구두점들
    #re.escape():파이썬 자체 내장 구두점들을 특수문자가 아닌 일반문자로 인식하도록
    text = re.sub(r'\w*\d\w*', '', text)
    #\w*\d\w*: 숫자 앞뒤로 단어가 붙어있는 경우 제거 예:room1,2nd,,,
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()

    return text
````

````python
#보류코드, 사용 X

nlp = spacy.load("en_core_web_sm") #spacy.load:영어문장을 분석할 수 있는 **전문가**를 데려오는 코드

def extract_noun_adj_pairs(text):
    pairs = []
    if not text or text.isspace():
        return pairs

    doc = nlp(text) #nlp가 text를 분석

    for token in doc: #token: 단어
        if token.pos_ == "ADJ" and token.head.pos_ in ["NOUN", "PROPN"]: #token.pos_: "이 단어의 품사는 무엇인가" #taken.head: token이 꾸며주는 단어
            pairs.append((token.text, token.head.text))


    return pairs #최종 반환
````

````python
# ==== 0) 준비: 라이브러리 ====
import os
import spacy
from tqdm import tqdm
import pandas as pd

# ==== 1) GPU 우선 사용 ====
# CuPy가 설치돼 있고 GPU가 가능하면 True가 떨어짐. 아니면 CPU로 자동 폴백.
if spacy.prefer_gpu():
    print("✅ Using GPU")
else:
    print("⚠️ GPU not available -> Using CPU")

# ==== 2) 모델 로드: 불필요한 컴포넌트 끄기 (속도↑) ====
nlp = spacy.load("en_core_web_sm", disable=["ner"])  # NER 비활성화
# 필요한 파이프만 더 줄이고 싶으면:
# nlp.select_pipes(enable=["tok2vec","tagger","parser","attribute_ruler","lemmatizer"])

# ==== 3) 헬퍼: 한 문서에서 (형용사 -> 명사/고유명사) 쌍 추출 ====
def extract_pairs_from_doc(doc):
    pairs = []
    for tok in doc:
        # 전형적인 '형용사→명사' 수식관계(amod)만 골라서 정확도+속도 ↑
        if tok.pos_ == "ADJ" and tok.dep_ == "amod" and tok.head.pos_ in ("NOUN", "PROPN"):
            # 가벼운 전처리: 소문자 표준화/표제어(lemma) 사용
            pairs.append((tok.lemma_.lower(), tok.head.lemma_.lower()))
    return pairs

# ==== 4) 입력 텍스트 준비 ====
low_score_df = df[df['Score_Difference'] < 0].copy()
texts_series = low_score_df['Cleaned_Negative_Review'].astype(str).str.strip()
texts = texts_series[texts_series != ""].tolist()  # tqdm ETA 정확도를 위해 리스트로

# ==== 5) 배치/병렬 파이프 + tqdm 진행바 ====
batch_size = 256
workers = max(1, (os.cpu_count() or 4) - 1)


````

````python
all_pairs = []
for doc in tqdm(nlp.pipe(texts, batch_size=batch_size, n_process=workers),
                total=len(texts), desc="Processing reviews"):
    all_pairs.extend(extract_pairs_from_doc(doc))

print(f"✅ Done. Extracted {len(all_pairs):,} (adj, noun) pairs.")

# ==== 6) 저장: 원시 쌍 + 빈도 집계 ====
pairs_df = pd.DataFrame(all_pairs, columns=["adj", "noun"])
pairs_df.to_csv("noun_adj_pairs_raw.csv", index=False, encoding="utf-8-sig")

freq_df = (pairs_df
           .value_counts(["adj","noun"])
           .rename("count")
           .reset_index()
           .sort_values("count", ascending=False))
freq_df.to_csv("noun_adj_pairs_freq.csv", index=False, encoding="utf-8-sig")

print("💾 Saved:")
print(" - noun_adj_pairs_raw.csv   (모든 쌍)")
print(" - noun_adj_pairs_freq.csv  (빈도 집계 상위부터)")
````

````python
import pandas as pd
from collections import Counter
import re
import string
import spacy
import os
from tqdm import tqdm

# 1. 데이터 로드 및 전처리
df = pd.read_csv('/content/drive/MyDrive/Hotel_Reviews.csv')
def clean_text(text):
    if not isinstance(text, str):
        return ''
    if 'no positive' in text.lower() or 'no negative' in text.lower():
        return ''
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()
    return text

# Positive_Review 컬럼을 전처리하여 Cleaned_Positive_Review 컬럼 생성
df['Cleaned_Positive_Review'] = df['Positive_Review'].apply(clean_text)

# Negative_Review 컬럼도 전처리
df['Cleaned_Negative_Review'] = df['Negative_Review'].apply(clean_text)
````

````python
from scipy import stats
from collections import Counter
import spacy

pairs_df = pd.read_csv('noun_adj_pairs_freq.csv')
positive_adjectives = [
    'great', 'nice', 'friendly', 'excellent', 'fantastic',
    'convenient', 'quick', 'perfect', 'spacious', 'helpful',
    'good', 'clean', 'comfortable', 'hot', 'free', 'next',
    'first', 'double', 'little', 'only', 'other', 'single',
    'superior', 'twin', 'old', 'real', 'much', 'enough', 'several'
]

# 2. 분석 대상 상위 20개 쌍 리스트 생성
top_20_pairs = [tuple(row) for row in most_common_filtered_pairs[['adj', 'noun']].values]


# 3. '핵심 쌍' 포함 여부 변수 생성 (전체 데이터에 적용)
def has_target_pair_efficient(text):
    if not isinstance(text, str):
        return False
    text_words = set(text.split())
    for adj, noun in top_20_pairs:
        if adj in text_words and noun in text_words:
            return True
    return False

# 긍정 리뷰 컬럼에 함수 적용
df['Has_Target_Pair'] = df['Cleaned_Positive_Review'].apply(has_target_pair_efficient)

# 4. 교차표 생성 및 카이제곱 검정 실행
# 'Reviewer_Score'와 'Average_Score' 컬럼을 바탕으로 평점 범주를 생성합니다.
df['Score_Category'] = df['Reviewer_Score'].apply(lambda x: 'High' if x >= df['Average_Score'].mean() else 'Low')

contingency_table = pd.crosstab(df['Has_Target_Pair'], df['Score_Category'])
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print("\n---")
print("### 카이제곱 검정 결과")
print(f"Chi-square Statistic (카이제곱 통계량): {chi2:.2f}")
print(f"p-value (p-값): {p_value:.10f}")
````

````python
#코드 반복
import pandas as pd
from collections import Counter
import re
import string
import spacy
import os
from tqdm import tqdm

# 1. 데이터 로드 및 전처리
df = pd.read_csv('/content/drive/MyDrive/Hotel_Reviews.csv')
def clean_text(text):
    if not isinstance(text, str):
        return ''
    if 'no positive' in text.lower() or 'no negative' in text.lower():
        return ''
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()
    return text

df['Cleaned_Positive_Review'] = df['Positive_Review'].apply(clean_text)

# 2. 명사-형용사 쌍 추출
nlp = spacy.load("en_core_web_sm")
def extract_pairs_from_doc(doc):
    pairs = []
    for tok in doc:
        if tok.pos_ == "ADJ" and tok.dep_ == "amod" and tok.head.pos_ in ("NOUN", "PROPN"):
            pairs.append((tok.lemma_.lower(), tok.head.lemma_.lower()))
    return pairs

# '높은 평점' 리뷰만 추출합니다.
high_score_df = df[df['Reviewer_Score'] >= df['Average_Score']].copy()
texts_series = high_score_df['Cleaned_Positive_Review'].astype(str).str.strip()
texts = texts_series[texts_series != ""].tolist()
batch_size = 256
workers = max(1, (os.cpu_count() or 4) - 1)

all_positive_pairs = []
for doc in tqdm(nlp.pipe(texts, batch_size=batch_size, n_process=workers),
                 total=len(texts), desc="Processing positive reviews"):
    all_positive_pairs.extend(extract_pairs_from_doc(doc))

# 3. 빈도수 계산 및 저장
positive_pairs_freq = Counter(all_positive_pairs).most_common()

positive_pairs_df = pd.DataFrame(positive_pairs_freq, columns=['adj_noun_pair', 'count'])
positive_pairs_df[['adj', 'noun']] = pd.DataFrame(positive_pairs_df['adj_noun_pair'].tolist())
positive_pairs_df = positive_pairs_df.drop(columns=['adj_noun_pair'])
positive_pairs_df = positive_pairs_df[['adj', 'noun', 'count']]
positive_pairs_df.to_csv("positive_noun_adj_pairs_freq.csv", index=False, encoding="utf-8-sig")
````

````python
import pandas as pd
from collections import Counter
import re
import string
import spacy
import os
from tqdm import tqdm

df['Cleaned_Positive_Review'] = df['Positive_Review'].apply(clean_text)
df['Cleaned_Negative_Review'] = df['Negative_Review'].apply(clean_text)

# 2. 명사, 형용사, 쌍 추출 함수 정의
nlp = spacy.load("en_core_web_sm")

def extract_adjectives_from_doc(doc):
    return [tok.lemma_.lower() for tok in doc if tok.pos_ == "ADJ"]

def extract_nouns_from_doc(doc):
    return [tok.lemma_.lower() for tok in doc if tok.pos_ in ("NOUN", "PROPN")]

def extract_pairs_from_doc(doc):
    pairs = []
    for tok in doc:
        if tok.pos_ == "ADJ" and tok.dep_ == "amod" and tok.head.pos_ in ("NOUN", "PROPN"):
            pairs.append((tok.lemma_.lower(), tok.head.lemma_.lower()))
    return pairs

# 3. 긍정/부정 형용사 리스트 정의
positive_adjectives = [
    'great', 'nice', 'friendly', 'excellent', 'fantastic', 'convenient',
    'quick', 'perfect', 'spacious', 'helpful', 'good', 'clean', 'comfortable'
]
negative_adjectives = [
    'small', 'uncomfortable', 'poor', 'noisy', 'bad', 'dirty',
    'tiny', 'rude', 'broken', 'slow', 'old'
]
````

````python
# 4. 리뷰별 데이터 추출 및 집계
positive_texts = df['Cleaned_Positive_Review'].astype(str).str.strip().tolist()
negative_texts = df['Cleaned_Negative_Review'].astype(str).str.strip().tolist()
workers = max(1, (os.cpu_count() or 4) - 1)

all_positive_pairs = []
all_positive_adjectives = []
all_positive_nouns = []
for doc in tqdm(nlp.pipe(positive_texts, n_process=workers), total=len(positive_texts), desc="Processing positive reviews"):
    all_positive_pairs.extend(extract_pairs_from_doc(doc))
    all_positive_adjectives.extend(extract_adjectives_from_doc(doc))
    all_positive_nouns.extend(extract_nouns_from_doc(doc))

all_negative_pairs = []
all_negative_adjectives = []
all_negative_nouns = []
for doc in tqdm(nlp.pipe(negative_texts, n_process=workers), total=len(negative_texts), desc="Processing negative reviews"):
    all_negative_pairs.extend(extract_pairs_from_doc(doc))
    all_negative_adjectives.extend(extract_adjectives_from_doc(doc))
    all_negative_nouns.extend(extract_nouns_from_doc(doc))
````

````python
# 1) 유니크한 명사/형용사 개수
unique_positive_nouns = len(set(all_positive_nouns))
unique_positive_adjectives = len(set(all_positive_adjectives))
unique_negative_nouns = len(set(all_negative_nouns))
unique_negative_adjectives = len(set(all_negative_adjectives))

# 2) 쌍 데이터프레임 생성
positive_pairs_df = pd.DataFrame(all_positive_pairs, columns=['adj', 'noun'])
negative_pairs_df = pd.DataFrame(all_negative_pairs, columns=['adj', 'noun'])

# 3) 필터링 조건에 따른 쌍 개수
# 부정에서 긍정 형용사 제거된 쌍
negative_filtered_by_positive = negative_pairs_df[~negative_pairs_df['adj'].isin(positive_adjectives)]
# 긍정에서 부정 형용사 제거된 쌍
positive_filtered_by_negative = positive_pairs_df[~positive_pairs_df['adj'].isin(negative_adjectives)]
````

````python
### 최종 결과 출력 ###
print("\n### 1. 고유한 명사 및 형용사 수 (긍정/부정)")
print(f"1) 긍정 리뷰의 고유한 명사 수: {unique_positive_nouns:,}")
print(f"2) 긍정 리뷰의 고유한 형용사 수: {unique_positive_adjectives:,}")
print(f"3) 부정 리뷰의 고유한 명사 수: {unique_negative_nouns:,}")
print(f"4) 부정 리뷰의 고유한 형용사 수: {unique_negative_adjectives:,}")

print("\n### 2. 필터링 전/후 쌍 개수")
print(f"5) 부정 리뷰의 모든 쌍 (제거되지 않음): {len(negative_pairs_df):,}")
print(f"6) 부정 리뷰에서 긍정 형용사 제거 후 쌍: {len(negative_filtered_by_positive):,}")
print(f"7) 긍정 리뷰의 모든 쌍 (제거되지 않음): {len(positive_pairs_df):,}")
print(f"8) 긍정 리뷰에서 부정 형용사 제거 후 쌍: {len(positive_filtered_by_negative):,}")
````

````python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('Hotel_Reviews.csv')

df['Score_Difference'] = df['Reviewer_Score'] - df['Average_Score']

# 중앙값 계산
median_score_diff = df['Score_Difference'].median()

# 절대 편차(MAD) 계산
mad = np.median(np.abs(df['Score_Difference'] - median_score_diff))

# 수정된 Z-점수(Modified Z-score) 계산
# MAD가 0인 경우를 방지하기 .. 0으로 나누는 오류를 막기
if mad == 0:
    df['Modified_Z_Score'] = 0
else:
    df['Modified_Z_Score'] = 0.6745 * (df['Score_Difference'] - median_score_diff) / mad

print("Score_Difference 통계량")
print(df['Score_Difference'].describe())
print("\n수정된 Z점수 통계량")
print(df['Modified_Z_Score'].describe())
````
